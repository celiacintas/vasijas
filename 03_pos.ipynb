```python
from glob import glob
import numpy as np
import pyvox.parser
from pyvox.models import Vox
from pyvox.writer import VoxWriter
```

```python
def __read_vox__(path):
        vox = pyvox.parser.VoxParser(path).parse()
        a = vox.to_dense()
        caja = np.zeros((64, 64, 64))
        caja[0:a.shape[0], 0:a.shape[1], 0:a.shape[2]] = a
        return caja
```

```python
vox = __read_vox__('model_BA_42_2-n003-t1649437795_Dilated.vox')
```

```python
a_part = np.zeros_like(vox, dtype=np.bool_)
a_part[31:32,0:32,0:64] = vox[31:32,0:32,0:64]
```

```python
a_part_vox = Vox.from_dense(a_part.reshape(64, 64, 64))
```

```python
VoxWriter(f'example.vox',a_part_vox).write()
```
