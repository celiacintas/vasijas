```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage.draw import circle
```

```python
import glob
```

```python
files = glob.glob("data/perfiles_CATA/png_clasificados/*/*.png")
```

```python
for f in files:
    print(f)
    img = cv2.imread(f, 0)
    img_f = np.ones(img.shape) * 255
    factor = img.shape
    factor_min = min(factor)
    factt = factor_min/3
    #plt.imshow(img)

    coords=[]
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if img[i][j]!= 255:
                coords.append((i,j))

    frac_cant = 5
    frac_step = int(len(coords) // frac_cant)
    #print(frac_cant, frac_step)
    #h,w = 100, 100
    h,w = int(factt), int(factt)
    
    for idx, coor in enumerate([coords[0], coords[len(coords) - 1]]):
    
        for i in range(frac_cant):
            #fig = plt.figure(figsize=(9, 6))
            #ax = fig.add_subplot(111)
            img_f = np.ones(img.shape) * 255
            x,y= coor
            #rr, cc = circle(x, y, 100, img.shape)
            rr, cc = circle(x, y, int(factt) * ((i + 1)*1.2), img.shape)
            h,w = int(int(factt) * ((i + 1)*1.2)), int(int(factt) * ((i + 1)*1.2))
            img_f[rr, cc] = img[rr, cc]
            #ax.imshow(img_f)
            lim = np.array([x-h,x+h,y-w,y+w])
            lim[lim<0] = 0
            lim[lim<0] = 0
            lim[0:2][lim[0:2]>img.shape[0]] = img.shape[0]
            lim[2:][lim[2:]>img.shape[1]] =  img.shape[1]
            crop_img = img_f[lim[0]:lim[1],lim[2]:lim[3]]
            frac_name_full = f.replace(".png", "_frac_{}_full.png".format(i+(idx * 5)))
            frac_name_part = f.replace(".png", "_frac_{}_part.png".format(i+(idx * 5)))
            #print(frac_name_part)
            cv2.imwrite(frac_name_full, img_f)
            cv2.imwrite(frac_name_part, crop_img)

            #ax.imshow(crop_img)
```
