```python
import numpy as np
import matplotlib.pyplot as plt
import glob
import cv2
```

```python
file_fake = glob.glob('imageFakeTest_v9/*')
```

```python
file_fake[0].replace('imageFakeTest_v9', 'parte')
```

```python
for f in file_fake:
    img = cv2.imread(f)
    img_parte = cv2.imread(f.replace('imageFakeTest_v9', 'parte'))
    plt.figure(figsize=(12,5))
    fig, (ax0) = plt.subplots(1, 1, figsize=(12,5))
    ax0.axis("off")
    ax0.imshow(img_parte)
    ax0.imshow(img, alpha=0.2)
    plt.xticks([])
    plt.yticks([])
    plt.savefig('superposicion_fake{}'.format(f.replace('imageFakeTest_v9', '')), bbox_inches='tight',  dpi=300)

```
