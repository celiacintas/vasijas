```python
# %%
import pandas as pd
# %%
import numpy as np
# %%
import glob
```

```python
# %%
df = pd.read_csv('data/porcentajes.csv')
```

```python
# %%
df['Fichero']
```

```python
# %%
lista_data = glob.glob('data/test/*/*')

# %%
vasijas_test = [f.split('/')[-1] for f in lista_data]

# %%
df_vasija_test =  df[df['Fichero'].isin(vasijas_test)]

# %%
for li,ls in [(1,15), (15,20), (20,30), (30,100)]:
    print('*'*20)
    total = df_vasija_test[np.logical_and(df_vasija_test[' percentage'] > li, ls>df_vasija_test[' percentage'])].shape[0]
    print(li, ls, total)
    aux_df = df_vasija_test[np.logical_and(df_vasija_test[' percentage'] > li, ls>df_vasija_test[' percentage'])].reset_index(drop=True)
    class_aux = [glob.glob(f'data/*/*/{i}')[0].split('/')[-2] for i in aux_df['Fichero'].tolist()]
    print('#'*10)
    class_aux = np.array(class_aux, np.int32)
    unique, counts = np.unique(class_aux, return_counts=True)
    for u,c in zip(unique, counts):
        p = round((c*100)/total, ndigits=2)
        print(f'    clase {u} -- {c} ({p}%)')

    #aux_df.to_csv(f'data/fragmentos_{li}_{ls}.csv')
```
