```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [9.0*1.5, 6.0*1.5]
```

```python
data = pd.read_csv("data/result.csv",index_col =0)
```

```python
files = np.array(data["image"])
```

```python
target = np.array(data["target"])
```

```python
X_net = np.array(data[data.columns[:-2]])
```

```python
def select_file(img_n):
    replace_string = re.findall("\/*_frac_\w_\w*.png", img_n)
    if len(replace_string)==0:
        replace_string =".png"
    else:
        replace_string = replace_string[0]
    img_select = img_n.replace(replace_string, "")
    return img_select
```

```python
img_n = files[idtest]
img_select = select_file(img_n)
```

```python
dist = []
for j,i in enumerate(X_net):
    a = scipy.spatial.distance.euclidean(i, X_net[idtest])
    img_n = files[j]
    img_n = select_file(img_n)
    if a == 0 or img_select==img_n:
        print(img_n, a)
        dist.append(100000000)
    else:
        dist.append(a)
```
